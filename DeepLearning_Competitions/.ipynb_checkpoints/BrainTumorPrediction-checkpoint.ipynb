```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

#ignore warnings
import logging
tf.get_logger().setLevel(logging.ERROR)
```

# Lets see the px counts of images to find lower_bound to convert    

<b> (300 x 300) can be appropriate lower bound

```python
train_data = keras.preprocessing.image_dataset_from_directory(
    directory= 'ImagesFolder/BrainTumor/train/',
    batch_size=32,
    seed=42,
    validation_split=0.01,
    image_size=(300,300),
    subset='training'
)

valid_data = keras.preprocessing.image_dataset_from_directory(
    directory= 'ImagesFolder/BrainTumor/valid/',
    batch_size=32,
    seed=42,
    validation_split=0.99,
    image_size=(300,300),
    subset='validation'
)
```

```python
train_data.class_names
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_df = train_data.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
valid_df = valid_data.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)

train_ds = train_df.map(lambda x, y: (tf.image.per_image_standardization(x), y))
valid_ds = valid_df.map(lambda x, y: (tf.image.per_image_standardization(x), y))
```

```python
#creating model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=100,input_shape=(300,300,3),kernel_size=(3,3),activation='relu'),
    tf.keras.layers.MaxPool2D((2,2)),
    
    tf.keras.layers.Conv2D(filters=100,kernel_size=(3,3),activation='relu'),
    tf.keras.layers.MaxPool2D((2,2)),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(500,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(300,activation='relu'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_ds,epochs=5)
```

```python
print(model.evaluate(train_ds))
print(model.evaluate(valid_ds))
```
