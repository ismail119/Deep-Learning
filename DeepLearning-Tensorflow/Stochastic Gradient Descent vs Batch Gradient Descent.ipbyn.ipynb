When we try to apply Gradient Descent to our data to find optimum equation for the story, if we have so many sample and feature, it means that we are going to make so many derivative calculations. To avoid this problem with big data problems we uses Stochastic Gradient Descent. The difference between them, in stochastic we pick up one sample from our dataset and adjust weights; whereas in normal gradient descent, we use all of data to adjust weights.So it provides speed and performance for our model.
