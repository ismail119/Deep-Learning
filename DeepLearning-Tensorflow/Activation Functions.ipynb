## Implementation of activation functions in python

<h3 style='color:purple'>Sigmoid</h3>

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
print("%.25f"%sigmoid(-56))
```

```python
sigmoid(0.5)
```

<h3 style='color:purple'>tanh</h3>

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-67)
```

```python
tanh(100)
```

```python
tanh(1)
```

<h3 style='color:purple'>ReLU</h3>

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python
relu(8)
```

<h3 style='color:purple'>Step</h3>

```python
def step(x):
    return 1 if x>0 else 0
```

```python
step(-2)
```

```python
step(100)
```
