<p style="color:blue;font-size:15px" >To find equations weights easily we use gradient descent method.<br>
It calculates metrics given by us such as mean absolute error, mean squared error, binary accurancy...<br>
It uses back propagation to optimize these metrics in each epoch.</p>
<br>
<b style="color:black;font-size:16px"> Important!!</b><br>
<pre style="font-size:16px;color:red"> 
It uses partial derivative to calculate changing effect of only one dependent variable(coefficiant). 
Then with back propagation it uses new coefficient for equation in each epoch. 
So in every new epoch we can say that the 
model try to make better its equation but if we teach so much with overdose epoch ,
it might be cause overfitting.</b>
