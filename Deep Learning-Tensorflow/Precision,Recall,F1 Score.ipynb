```python
#formulas and explainations

def recall(TP,FN):
    print("it returns true positive predictions / true positive + false negative,so in this equation mathematically, we focus on FN numbers.If recall is small then we have many FN.")
    return TP/(TP+FN)
```

```python
def precision(TP,TN):
    print("it returns true positive predictions / all true predictions,here in this equation mathematically, we focus on TP numbers.If precision is small then we have less TP.")
    return TP/TP+TN
```

```python
def f1_score(TP,FP,FN):
    print("we target to use all arguments with this equation. And again it generally focus on TP.")
    return TP/(TP+(FP+FN)/2)
```

```python
#for our spesific case:
recall(4,2)
```

```python
print(precision(4,1))
```

```python
print(f1_score(4,3,2))
```
